| Activity | Time | Description |
|---|---|---|
| Context | 3 min | "You're going to use AI tools -- let's make sure you use them well." Normalize it, remove the stigma, set the tone. |
| Core Material | 15 min | Three tiers of AI use: **Ask** (explain this error, what does this code do, give me a simpler example), **Verify** (is my logic right, what edge cases am I missing), **Generate** (write this for me). First two are encouraged. Third is a crutch at this stage -- it skips the learning. Show 2-3 live examples of good prompts vs lazy prompts and the difference in output quality. |
| Discussion | 5 min | "When this semester is over, the AI tools will still be there but the course won't. What are you actually paying for?" Let them arrive at the answer: you're building the judgment to know whether AI output is correct. You can't evaluate what you don't understand. |
| Takeaways | 2 min | One-page reference card: 10 starter prompts for learning (e.g., "Explain this error like I'm a beginner," "Show me three ways to do X," "What's wrong with this code?"). |
