### Variables Persist Across Cells

Variables defined in one cell are available in all subsequent cells (as long as you run them in order).

````python
# Define variables here
beam_length = 10.0  # meters
load = 5000         # Newtons
E = 200e9           # Elastic modulus (Pa)
I = 8.33e-6         # Moment of inertia (m^4)
````

````python
# Use them here
max_deflection = (load * beam_length**3) / (48 * E * I)
print(f"Max deflection: {max_deflection*1000:.2f} mm")
````

### Matplotlib Plots (Inline)

````python
import matplotlib.pyplot as plt
import numpy as np

# Beam deflection curve
x = np.linspace(0, beam_length, 100)
w = load  # point load at center
deflection = np.where(
    x <= beam_length/2,
    (w * x) / (48 * E * I) * (3 * beam_length**2 - 4 * x**2),
    (w * (beam_length - x)) / (48 * E * I) * (3 * beam_length**2 - 4 * (beam_length - x)**2)
)

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, deflection * 1000, color='#1e3a5f', linewidth=2)
ax.fill_between(x, deflection * 1000, alpha=0.1, color='#1e3a5f')
ax.set_xlabel('Position along beam (m)')
ax.set_ylabel('Deflection (mm)')
ax.set_title('Simply Supported Beam -- Deflection Curve')
ax.grid(True, alpha=0.3)
ax.invert_yaxis()
plt.tight_layout()
plt.show()
````

### Interactive Plots with Plotly

````python
import plotly.graph_objects as go

materials = ['Steel', 'Concrete', 'Aluminum', 'Wood', 'Glass']
strength = [250, 30, 270, 40, 33]
density = [7850, 2400, 2700, 500, 2500]

fig = go.Figure(data=go.Scatter(
    x=density,
    y=strength,
    mode='markers+text',
    text=materials,
    textposition='top center',
    marker=dict(size=15, color='#1e3a5f')
))
fig.update_layout(
    title='Material Strength vs Density',
    xaxis_title='Density (kg/m3)',
    yaxis_title='Tensile Strength (MPa)',
    template='plotly_white',
    width=700, height=400
)
fig.show()
````

### Pattern: Bug Hunt

Present broken code and ask students to find and fix the issue.

#### Bug Hunt

The following code should calculate the average of a list, but it has **two bugs**. Find and fix them.

````python
# BUGGY CODE -- find and fix the 2 bugs
def average(numbers):
    total = 0
    for i in range(len(numbers)):
        total += i                    # Bug 1: adding index, not value
    return total / len(numbers) + 1   # Bug 2: off-by-one in denominator

test_data = [10, 20, 30, 40, 50]
print(f"Average: {average(test_data)}")  # Should be 30.0
````

### Pattern: Side-by-Side Comparison

Show two approaches to the same problem.

````python
import time

# Approach 1: Loop
start = time.time()
total_loop = 0
for i in range(1_000_000):
    total_loop += i
time_loop = time.time() - start

# Approach 2: Built-in
start = time.time()
total_builtin = sum(range(1_000_000))
time_builtin = time.time() - start

print(f"Loop:     {total_loop:>15,} in {time_loop:.4f}s")
print(f"Built-in: {total_builtin:>15,} in {time_builtin:.4f}s")
print(f"Speedup:  {time_loop/time_builtin:.1f}x")
````

## 5. VS Code-Specific Features

These features work specifically in VS Code's Jupyter implementation.

### Keyboard Shortcuts

| Action | Shortcut |
|--------|----------|
| Run cell | `Shift + Enter` |
| Run cell, stay in place | `Ctrl + Enter` |
| Add cell above | `A` (in command mode) |
| Add cell below | `B` (in command mode) |
| Delete cell | `DD` (in command mode) |
| Toggle markdown/code | `M` / `Y` (in command mode) |
| Command mode | `Escape` |
| Edit mode | `Enter` |
| Toggle line numbers | `L` (in command mode) |

> **Command mode** = cell is selected (blue border) but you're not typing in it. Press `Escape` to enter command mode.

### Other VS Code Features

- **Outline panel**: Auto-generates a table of contents from markdown headers
- **Variable explorer**: Click the variables icon in the toolbar to inspect all active variables
- **Data viewer**: Click on a DataFrame or array in the variable explorer to view it as a table
- **Cell collapsing**: Click the arrow next to any cell to collapse it
- **Diff view**: Git integration shows notebook diffs cell-by-cell
- **IntelliSense**: Full autocomplete works in code cells
- **Debugging**: You can set breakpoints and debug individual cells

## 6. Notebook Best Practices for Lessons

1. **Run all cells top-to-bottom** before sharing -- ensures everything works in sequence
2. **Clear all outputs** before committing to Git -- keeps diffs clean (Ctrl+Shift+P > "Clear All Outputs")
3. **One concept per section** -- mirrors your modular block structure
4. **Use markdown cells liberally** -- the notebook IS the lesson, not a code dump
5. **Name variables descriptively** -- students read your code as a model
6. **Include expected output** in markdown -- so students know what success looks like even if they can't run the cell yet
